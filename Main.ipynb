## Model Building 

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
df_model
```

```python
df_model.drop(columns=['ID'])
```

```python
df_model['Income'] = df_model['Income'].fillna(df_model['Income'].median())
```

```python
X = df_model[[
    "Income",
    "Recency",
    "MntWines",
    "MntMeatProducts",
    "MntFishProducts",
    "MntSweetProducts",
    "NumWebPurchases",
    "NumStorePurchases",
]]

```

```python
print(X.isnull().sum())
```

```python
# For every value of K, K-Means gives a WCSS (Inertia) value.

wcss = []
```

```python
wcss = []

for i in range(1,11):

    kmeans = KMeans(
        n_clusters=i,
        random_state=42
    )

    kmeans.fit(X)

    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.Figure(figsize=(9,6))
plt.plot(range(1,11),wcss,marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.grid()
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X)
```

```python
df_model['Cluster'] = clusters
```

```python
df_model.head()
```

```python
df_model['Cluster'].value_counts()
```

```python
df_model.groupby('Cluster').mean()
```

```python
important_features = ['Income',
        'Recency',
        'MntWines',
        'MntMeatProducts',
        'MntFishProducts',
        'MntSweetProducts',
        'NumWebPurchases',
        'NumStorePurchases']

df_model.groupby('Cluster')[important_features].mean()
```
