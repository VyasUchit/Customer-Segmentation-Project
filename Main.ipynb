## One-Hot Encoder :

```python
df_model = pd.get_dummies(df_model, columns=['Education','Marital_Status'],dtype=int)
```

```python
df_model.head()
```

```python
df_model = df_model.dropna(subset=["Income"])

print(df_model["Income"].isnull().sum())
```

## Standard Scalling 

```python
features = [
    "Income",
    "Recency",
    "MntWines",
    "MntMeatProducts",
    "MntFishProducts",
    "MntSweetProducts",
    "NumWebPurchases",
    "NumStorePurchases"
]

X = df_model[features]
```

```python
print(X.isnull().sum())
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
df_model.drop(columns=['ID'])
```

```python
from sklearn.cluster import KMeans

wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.Figure(figsize=(9,6))
plt.plot(range(1,11),wcss,marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.grid()
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

df_model["Cluster"] = clusters
```

```python
df_model.head()
```

```python
df_model['Cluster'].value_counts()
```

```python
df_model.groupby("Cluster")[[
    "Income",
    "Recency",
    "MntWines",
    "MntMeatProducts",
    "MntFishProducts",
    "MntSweetProducts",
    "NumWebPurchases",
    "NumStorePurchases"
]].mean()
```

```python
df_model['Cluster'].value_counts()
```

```python
df_model['Cluster'].value_counts().plot(kind='bar')
plt.title("Customer Segments")
plt.xlabel("Cluster")
plt.ylabel("Number of Customers")
plt.show()
```

```python
# Number of customers in each cluster

cluster_counts = df_model['Cluster'].value_counts().sort_index()

plt.bar(cluster_counts.index, cluster_counts.values)
plt.title("Customer Segments")
plt.xlabel("Cluster")
plt.ylabel("Number of Customers")
plt.show()
```
